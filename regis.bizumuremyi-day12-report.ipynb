# Practical_7: Matplotlib

````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
# Plot a sine wave using Matplotlib for x values between 0 and 2π.
x=np.linspace(0,2*np.pi)
y=np.sin(x)
plt.plot(x,y,linewidth=2,c="r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sine wave");
````

````python
# Create a bar chart for the following data: categories = ['A', 'B', 'C', 'D'] and values = [3, 7, 5, 9].
categories=['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]
plt.bar(categories,values,color ='maroon', width = 0.4)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar chart for Categories and Values");
````

````python
# Create a histogram of 1000 random numbers generated from a normal distribution.
x=np.random.normal(0,1,1000)
plt.hist(x,color="green")
plt.xlabel("X")
plt.ylabel("Frequence")
plt.title("Normal distribution");
````

````python
# Plot two line plots on the same graph: y1 = x^2 and y2 = x^3 for x ranging from 0 to 10.
x=np.arange(11)
y1=x**2
y2=x**3
plt.figure(figsize=(10,8))
plt.plot(y1,c="b")
plt.plot(y2,c="brown");
````

````python
#Add titles, axis labels, and a legend to the plot in Exercise 7.
x=np.arange(11)
y1=x**2
y2=x**3
plt.figure(figsize=(10,8))
plt.plot(y1,c="b",label="y1")
plt.plot(y2,c="brown",label="y2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Combination of  $y_1=x^2$ and $y_2=x^3$")
plt.legend();
````

````python
# Create a pie chart with 5 segments labeled A, B, C, D, and E, with values [25, 35, 20, 15, 5].
values=np.array([25, 35, 20, 15, 5])
label=["A","B","C","D","E"]
plt.pie(values,labels=label)
plt.title("Pie Chart");
````

````python
#Create a subplot with 2 rows and 1 column. Plot y1 = x^2 in the first subplot and y2 = x^3 in the second.
x=np.arange(11)
y1=x**2
y2=x**3
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(y1)
axes[1].plot(y2);
````

````python
# Use a different color and linestyle for each line in a plot of three functions: y1 = x, y2 = x^2, and y3 = x^3.
x=np.arange(11)
y=x
y1=x**2
y2=x**3
plt.plot(y,c="r",linestyle="--")
plt.plot(y1,c="brown",ls="solid")
plt.plot(y2,c="g",ls=":")
````
